    - 클래스와 인스턴스는 다른 객체들과의 상호작용에서 서로 충돌이나 영향을 주지 않으면서 독립적으로 동작할 수 있음
    - 코드의 가독성, 유지보수성, 재사용성을 높이는데 도움을 줌






```python
# 클래스 정의
class Person:
    # 속성(변수)
    count = 0
    blood_color = 'red' # 클래스 변수 - 클래스 내부에 선언된 변수
    #클래스로 생성된 모든 인스턴스들이 공유하는 변수

    #메서드 
    def __init__(self, name):
        self.name = name
        Person.count += 1
    #생성될 때 자동으로 호출되므로 이렇게 활용도 가능
    #객체를 생성할 때 무슨 동작을 하고 싶으면 코드를 작성해야됨
    
    # __init__는 매직매서드로 개발자가 직접호출x 알아서 호출됨
    # 객체의 초기화를 담당, ,생성자 함수를 통해 인스턴스를 생성하고 필요한 초기값을 설정
    # 생성자 메서드
    # 인스턴스를 생성될 때 데이터를 초기화 시켜줘야 되므로 써주는 것


    #인스턴스 변수
    ## 인스턴스(객체)마다 별도로 유지되는 변수
    ## 인스턴스마다 독립적인 값을 가지며 인스턴스가 사용될 때마다 초기화 됨

    
    
    def singing(self):
        return f'{self.name}가 노래합니다.'
    #인스턴스 메서드
        # 각각의 인스턴스에서 호출할 수 있는 메서드

# 인스턴스 생성
singer1 = Person('iu') # name이 위치인자이므로 값을 넣어줘야 됨
singer2 = Person('BTS') # 똑같은 클래스로 만든 다른 인스턴스

# 메서드 호출 return이 있어 print함수 써줌
# 속성중심인것을 알 수 있음
print(singer1.singing()) 
print(singer2.singing())

print(Person.count)
# 속성(변수)
print(singer1.blood_color)
print(singer2.blood_color)

#인스턴스와 클래스 간의 이름 공간
#인스턴스와도 클래스와도 name은 공유하지 않음 - 크래스 변수만 공유


    
```

##### 메서드 종류

- 메서드는 행동을 정의

1. 인스턴스 메서드
    - 인스턴스.메서드 - 인스턴스가 쓰는 메서드
    - 클래스로부터 생성된 각 인스턴스에서 호출할 수 있는 메서드 - 인스턴스의 상태를 조작하거나 동작을 수행 [1,2,3].append() - append는 인스턴스 메서드
    - 클래스 내부에 정의되는 메서드의 기본
    - 반드시 첫 번째 매개변수로 인스턴스 자신(self)을 전달받음
    - 매개변수 self를 다른이름으로 만들어도 아무상관 x but. 쓸수있지만 안씀
    - 각 인스턴스변수에 접근하고 결과가 달라질 수 있음
    


2. 클래스 메서드
    - 클래스.메서드 - 클래스가 쓰는 메서드
        - 대부분은 인스턴스가 하지만 수정하기 위해 클래스도 메서드 있을 수 있음
        - 클래스가 호출하는 메서드
        - 클래스 변수를 조작하거나 클래스 레벨의 동작을 수행
        - 위쪽에 @classmethod(데코레이터)붙어있고 첫번째 변수가 self가 아닌 cls가 있음
        - class_method가 호출하는 함수가 첫번째 변수로 들어감
        - @classmethod이게 있어야 class에서 동작함 인스턴스에서 동작x 무언가의 함수를 꾸며주는 역할

3. 정적 메서드(static method)
    - 클래스와 인스턴스와 상관없이 독립적으로 동작하는 메서드
    - **주로 클래스와 관련이 있지만 인스턴스와 상호작용** - 클래스로 수학의 함수를 모아놓는 다면 안에 정적 메서드 여러개 있을 수 있음 클래스와 인스턴스의 데이터 조작은 X
    - 데코레이터 @staticmethod적어줌
    - *호출 시 필수적으로 작성해야 할 매개변수가 없음*
    - 즉, 객체 상태나 클래스 상태를 수정할 수 없으며 단지 기능(행동)만을 위한 메서드로 사용
    - 클래스, 인스턴스 이런거 신경안쓰고 걍 어떠한 기능에 초점을 맞춘 것

```python
# 인스턴스.메서드()
'abc'.upper() # 객체가 뭔가를 함 즉 self를 쓰는 이유
# upper
# 클래스.메서드(인스턴스 자기자신) 위에껀 축약형
str.upper('abc') # 객체지향적인 코드는 아님 데이터가 그냥 인자로서 들어감
# str클래스가 upper 메서드를 호출했고,
# 그 첫번째 인자로 문자열 인스턴스가 들어간 것
```

```python
class Shape:

    def __init__(self, w, h):
        self.w = w
        self.h = h

    def __gt__(self, other):
        return self.w * self.h > other.w * other.h

    def __eq__(self, other):
        return self.w * self.h == other.w * other.h



s1 = Shape(3,4)
s2 = Shape(5,6)

print(s2 > s1)

s3 = Shape(6,5)

print(s2 == s1)
```
